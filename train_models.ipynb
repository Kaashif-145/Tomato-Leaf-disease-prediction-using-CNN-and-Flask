```python
# 📌 1. Imports & Config 
import os
import sys
import numpy as np
import tensorflow as tf
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import classification_report, confusion_matrix
from tensorflow.keras import layers, regularizers
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.applications import (
    EfficientNetB0, NASNetMobile, EfficientNetV2S, DenseNet121
)
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.callbacks import EarlyStopping, ReduceLROnPlateau
from tensorflow.keras.models import Model
from utils22 import build_model  # Ensure this supports dropout or modify it below

# Config
IMG_SIZE = (224, 224)
BATCH_SIZE = 32
EPOCHS = 20
DATA_DIR = r"D:\WORK\train"
NUM_CLASSES = 10
SELECTED_CLASSES = [
    "Tomato___Bacterial_spot",
    "Tomato___Early_blight",
    "Tomato___Late_blight",
    "Tomato___Leaf_Mold",
    "Tomato___Septoria_leaf_spot",
    "Tomato___Spider_mites Two-spotted_spider_mite",
    "Tomato___Target_Spot",
    "Tomato___Tomato_Yellow_Leaf_Curl_Virus",
    "Tomato___Tomato_mosaic_virus",
    "Tomato___healthy"
]
 # (same as before)

# 📌 2. Data Generators
train_aug = ImageDataGenerator(
    rescale=1./255, validation_split=0.2,
    rotation_range=30, width_shift_range=0.2, height_shift_range=0.2,
    shear_range=0.15, zoom_range=0.2, horizontal_flip=True
)

train_gen = train_aug.flow_from_directory(
    DATA_DIR,
    target_size=IMG_SIZE,
    batch_size=BATCH_SIZE,
    class_mode='categorical',
    subset='training',
    classes=SELECTED_CLASSES,
    shuffle=True
)

val_gen = train_aug.flow_from_directory(
    DATA_DIR,
    target_size=IMG_SIZE,
    batch_size=BATCH_SIZE,
    class_mode='categorical',
    subset='validation',
    classes=SELECTED_CLASSES,
    shuffle=False
)

# Callbacks
early_stop = EarlyStopping(monitor='val_loss', patience=3, restore_best_weights=True)
lr_reduce = ReduceLROnPlateau(monitor='val_loss', factor=0.2, patience=2, verbose=1)

# 📌 3. Updated Model Builder (dropout + regularization)
def build_model(base_model, input_shape, num_classes, fine_tune_at=100):
    base_model.trainable = True
    if fine_tune_at:
        for layer in base_model.layers[:fine_tune_at]:
            layer.trainable = False

    inputs = layers.Input(shape=input_shape)
    x = base_model(inputs, training=False)
    x = layers.GlobalAveragePooling2D()(x)
    x = layers.Dropout(0.5)(x)  # Added dropout
    x = layers.Dense(256, activation='relu', kernel_regularizer=regularizers.l2(0.01))(x)  # L2 regularization
    x = layers.Dropout(0.3)(x)
    outputs = layers.Dense(num_classes, activation='softmax')(x)
    return Model(inputs, outputs)

# 📌 4. Training Function with Metrics & Plots
def train_model(model_name, model_obj):
    print(f"\n🚀 Training {model_name}...\n")
    model = build_model(model_obj, IMG_SIZE + (3,), NUM_CLASSES, fine_tune_at=100)
    model.compile(optimizer=Adam(1e-4), loss='categorical_crossentropy', metrics=['accuracy'])

    history = model.fit(
        train_gen,
        validation_data=val_gen,
        epochs=EPOCHS,
        callbacks=[early_stop, lr_reduce]
    )

    model.save(f"{model_name.replace(' ', '_').lower()}_plant_model.h5")
    print(f"✅ {model_name} training complete.")

    val_loss, val_acc = model.evaluate(val_gen, verbose=0)
    print(f"📊 Final Validation Accuracy: {val_acc * 100:.2f}%")
    print(f"📉 Final Validation Loss: {val_loss:.4f}")

    # Accuracy & Loss Plot
    plt.figure(figsize=(12, 5))
    plt.subplot(1, 2, 1)
    plt.plot(history.history['accuracy'], label='Train Acc')
    plt.plot(history.history['val_accuracy'], label='Val Acc')
    plt.title(f'{model_name} Accuracy')
    plt.xlabel('Epochs'); plt.ylabel('Accuracy'); plt.legend()

    plt.subplot(1, 2, 2)
    plt.plot(history.history['loss'], label='Train Loss')
    plt.plot(history.history['val_loss'], label='Val Loss')
    plt.title(f'{model_name} Loss')
    plt.xlabel('Epochs'); plt.ylabel('Loss'); plt.legend()
    plt.tight_layout()
    plt.show()

    # Confusion Matrix & Classification Report
    val_gen.reset()
    y_pred = model.predict(val_gen)
    y_true = val_gen.classes
    y_pred_classes = np.argmax(y_pred, axis=1)

    print("\n📄 Classification Report:\n")
    print(classification_report(y_true, y_pred_classes, target_names=SELECTED_CLASSES))

    cm = confusion_matrix(y_true, y_pred_classes)
    plt.figure(figsize=(10, 8))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=SELECTED_CLASSES, yticklabels=SELECTED_CLASSES)
    plt.title(f'{model_name} Confusion Matrix')
    plt.xlabel('Predicted')
    plt.ylabel('True')
    plt.xticks(rotation=45, ha='right')
    plt.yticks(rotation=0)
    plt.tight_layout()
    plt.show()

# 📌 5. Models Dictionary
MODELS = {
    "DenseNet121": DenseNet121(include_top=False, weights='imagenet', input_shape=IMG_SIZE + (3,)),
    "EfficientNetV2S": EfficientNetV2S(include_top=False, weights='imagenet', input_shape=IMG_SIZE + (3,)),
    "EfficientNetB0": EfficientNetB0(include_top=False, weights='imagenet', input_shape=IMG_SIZE + (3,)),
    "NASNetMobile": NASNetMobile(include_top=False, weights='imagenet', input_shape=IMG_SIZE + (3,)),
}

# 📌 6. Train All Models
for name, base_model in MODELS.items():
    if base_model is not None:
        train_model(name, base_model)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from skimage.segmentation import mark_boundaries
from lime import lime_image
from tensorflow.keras.preprocessing import image
from tensorflow.keras.models import load_model

# ----------------------
# 1. Load your model
# ----------------------
model = load_model(r"D:\WORK\nasnetmobile_plant_model.h5")  # Change path to your saved model

# ----------------------
# 2. Define class names (must match your training order)
# ----------------------
class_names = [
    "Tomato___Bacterial_spot",
    "Tomato___Early_blight",
    "Tomato___Late_blight",
    "Tomato___Leaf_Mold",
    "Tomato___Septoria_leaf_spot",
    "Tomato___Spider_mites_Two_spotted_spider_mite",
    "Tomato___Target_Spot",
    "Tomato___Tomato_Yellow_Leaf_Curl_Virus",
    "Tomato___Tomato_mosaic_virus",
    "Tomato___healthy"
]


# ----------------------
# 3. Initialize LIME explainer
# ----------------------
explainer = lime_image.LimeImageExplainer()

# ----------------------
# 4. Prediction function for LIME
# ----------------------
def predict_fn(images):
    images = np.array(images)
    return model.predict(images)

# ----------------------
# 5. Function to load and preprocess images
# ----------------------
def load_and_preprocess(img_path, target_size=(224, 224)):
    img = image.load_img(img_path, target_size=target_size)
    img_array = image.img_to_array(img)
    img_array = np.expand_dims(img_array, axis=0) / 255.0
    return img_array[0], img_array

# ----------------------
# 6. Image paths
# ----------------------
image_paths = [
    r"D:\WORK\train\Tomato___Early_blight\0abc57ec-7f3b-482a-8579-21f3b2fb780b___RS_Erly.B 7609.JPG",
    r"D:\WORK\train\Tomato___Septoria_leaf_spot\0a68a294-30d1-4422-ab7e-a1909ec277f7___JR_Sept.L.S 8443.JPG",
    r"D:\WORK\train\Tomato___healthy\ffb39943-eabb-42cf-ad09-b17019e46d66___RS_HL 9871.JPG"
]

# ----------------------
# 7. Create figure for all results
# ----------------------
# Create figure for all results
# Create figure for all results
fig, axes = plt.subplots(len(image_paths), 3, figsize=(12, 12))

# Add title further above images
fig.suptitle("LIME Visualizations for Three Categories", fontsize=18, weight='bold')

# Adjust layout so title is higher
plt.subplots_adjust(top=0.90)  # Increase gap between title and images



for i, path in enumerate(image_paths):
    # Load and preprocess image
    img_disp, img_input = load_and_preprocess(path)

    # Get model prediction
    pred_probs = model.predict(img_input)
    pred_class = np.argmax(pred_probs)
    pred_label = class_names[pred_class]

    # Generate LIME explanation
    explanation = explainer.explain_instance(
        img_disp.astype('double'),
        predict_fn,
        top_labels=1,
        hide_color=0,
        num_samples=1000
    )

    temp, mask = explanation.get_image_and_mask(
        label=explanation.top_labels[0],
        positive_only=True,
        hide_rest=False,
        num_features=10,
        min_weight=0.0
    )

    # Plot Original Image
    axes[i, 0].imshow(img_disp)
    axes[i, 0].set_title("Original")
    axes[i, 0].axis('off')

    # Plot Prediction
    axes[i, 1].imshow(img_disp)
    axes[i, 1].set_title(f"Prediction: {pred_label}")
    axes[i, 1].axis('off')

    # Plot LIME Heatmap
    axes[i, 2].imshow(mark_boundaries(img_disp, mask, color=(1, 1, 0)))
    axes[i, 2].set_title("LIME Heatmap")
    axes[i, 2].axis('off')

plt.tight_layout()
plt.savefig(r"D:\WORK\Figure_6.png", dpi=300)  # Save for paper
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.models import load_model
from sklearn.metrics import classification_report, confusion_matrix

# Paths
test_dir = r'D:\WORK\test'
model_paths = {
    "EfficientNetB0": "efficientnetb0_plant_model.h5",
    "EfficientNetV2S": "efficientnetv2s_plant_model.h5",
    "NASNetMobile": "nasnetmobile_plant_model.h5",
    "DenseNet121": "densenet121_plant_model.h5"
}

# Config
IMG_SIZE = (224, 224)
BATCH_SIZE = 32

# Test Generator
test_datagen = ImageDataGenerator(rescale=1./255)
test_generator = test_datagen.flow_from_directory(
    test_dir,
    target_size=IMG_SIZE,
    batch_size=BATCH_SIZE,
    class_mode='categorical',
    shuffle=False
)

# Class labels
class_indices = test_generator.class_indices
class_names = list(class_indices.keys())

#  Evaluate Each Model
for model_name, path in model_paths.items():
    print(f"\n🔍 Evaluating: {model_name}")
    model = load_model(path)

    # Loss & Accuracy
    loss, accuracy = model.evaluate(test_generator, verbose=0)
    print(f"✅ Accuracy: {accuracy * 100:.2f}% | ❌ Loss: {loss:.4f}")

    # Predictions
    y_pred_probs = model.predict(test_generator, verbose=0)
    y_pred = np.argmax(y_pred_probs, axis=1)
    y_true = test_generator.classes

    # Classification Report
    print("\n📄 Classification Report:")
    print(classification_report(y_true, y_pred, target_names=class_names))

    #  Confusion Matrix
    cm = confusion_matrix(y_true, y_pred)
    plt.figure(figsize=(10, 8))
    sns.heatmap(cm, annot=True, fmt='d', cmap='YlGnBu', xticklabels=class_names, yticklabels=class_names)
    plt.title(f'{model_name} Confusion Matrix')
    plt.xlabel('Predicted')
    plt.ylabel('Actual')
    plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import roc_curve, auc
from sklearn.preprocessing import label_binarize
from tensorflow.keras.models import load_model
from keras.preprocessing.image import ImageDataGenerator

# Define image size and path
IMG_SIZE = (224, 224)
BATCH_SIZE = 32
VAL_DIR = r"D:\WORK\valid"  # Path to your validation data folder

# Create validation data generator
val_datagen = ImageDataGenerator(rescale=1./255)
val_generator = val_datagen.flow_from_directory(
    VAL_DIR,
    target_size=IMG_SIZE,
    batch_size=BATCH_SIZE,
    class_mode='categorical',
    shuffle=False
)

# Load the model
model = load_model(r"D:\WORK\nasnetmobile_plant_model.h5")

# Extract true labels and class information
y_true = val_generator.classes
class_names = list(val_generator.class_indices.keys())
n_classes = len(class_names)

# Predict class probabilities
y_pred_proba = model.predict(val_generator, verbose=1)

# Binarize true labels
y_true_bin = label_binarize(y_true, classes=np.arange(n_classes))

# Compute ROC curve and AUC for each class
fpr = dict()
tpr = dict()
roc_auc = dict()

for i in range(n_classes):
    fpr[i], tpr[i], _ = roc_curve(y_true_bin[:, i], y_pred_proba[:, i])
    roc_auc[i] = auc(fpr[i], tpr[i])

# Plot all ROC curves
plt.figure(figsize=(12, 9))
colors = plt.cm.get_cmap('tab10', n_classes)

for i in range(n_classes):
    plt.plot(fpr[i], tpr[i], color=colors(i), lw=2,
             label=f"{class_names[i]} (AUC = {roc_auc[i]:.2f})")

plt.plot([0, 1], [0, 1], 'k--', lw=2)
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel("False Positive Rate", fontsize=13)
plt.ylabel("True Positive Rate", fontsize=13)
plt.title("Multi-Class ROC Curve - NASNetMobile", fontsize=15)
plt.legend(loc="lower right", fontsize=10)
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
from tensorflow.keras.models import load_model
from tensorflow.keras.preprocessing.image import ImageDataGenerator
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import precision_recall_curve, average_precision_score
from sklearn.preprocessing import label_binarize

# 1. Load models
nasnet_model = load_model(r"D:\WORK\nasnetmobile_plant_model.h5")
densenet_model = load_model(r"D:\WORK\densenet121_plant_model.h5")
efficientnet_model = load_model(r"D:\WORK\efficientnetv2s_plant_model.h5")

# 2. Load test data
test_datagen = ImageDataGenerator(rescale=1./255)
test_generator = test_datagen.flow_from_directory(
    r"D:\WORK\test",
    target_size=(224, 224),
    batch_size=32,
    class_mode='categorical',
    shuffle=False
)

# 3. True labels
y_true = test_generator.classes
n_classes = len(np.unique(y_true))
y_true_bin = label_binarize(y_true, classes=np.arange(n_classes))

# 4. Function to compute PR curve data
def get_pr_curve_data(model, model_name):
    y_pred_probs = model.predict(test_generator)
    precision, recall, _ = precision_recall_curve(y_true_bin.ravel(), y_pred_probs.ravel())
    ap = average_precision_score(y_true_bin, y_pred_probs, average="micro")
    return precision, recall, ap, model_name

# 5. Collect PR curve data
pr_data = []
pr_data.append(get_pr_curve_data(nasnet_model, "NASNetMobile"))
pr_data.append(get_pr_curve_data(densenet_model, "DenseNet121"))
pr_data.append(get_pr_curve_data(efficientnet_model, "EfficientNetV2S"))

# 6. Plot
plt.figure(figsize=(8, 6))
for precision, recall, ap, name in pr_data:
    plt.plot(recall, precision, lw=2, label=f"{name} (AP = {ap:.2f})")

plt.xlabel("Recall")
plt.ylabel("Precision")
plt.title("Precision–Recall Curves (Micro-Average) for All Models")
plt.legend(loc="best")
plt.grid(True)
plt.show()
```
